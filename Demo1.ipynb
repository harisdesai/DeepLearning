```python
import numpy as np

np.random.seed(42)

layers = 10

weights = []
biases = []

for i in range(layers):
  w = np.random.randn(1,1)
  b = np.random.uniform(1)
  weights.append(w)
  biases.append(b)

def relu(x):
    return np.maximum(0, x)

# def tanh(x):
#   return np.tanh(x)

def sigmoid(x):
  return 1 / (1 + np.exp(-x))

def forward(x):
  for i in range(layers):
    x= np.dot(x, weights[i]) + biases[i]
    # x = sigmoid(x)
    x = relu(x)
    # x = tanh(x)
    print(x)
  return x

x = np.array([[1.0]])

output = forward(x)
print("Output of DeepLearning Model:", output)
```

**Vanishing** **Gradient**

```python
weight_vanish = [np.random.randn(1,1)*0.01 for _ in range(layers)]

for w in weight_vanish:
  x = np.dot(x,w)
  print(x)
  x = sigmoid(x)

print("Final Output",x)
```

**Eploding** **Gradient**

```python
weight_vanish = [np.random.randn(1,1)*5 for _ in range(layers)]

for w in weight_vanish:
  x = np.dot(x,w)
  print(x)
  x = sigmoid(x)

print("Final Output",x)
```

**Symmetry Problem**

```python
x = np.array([[10,5]])

w1 = np.array([0.5,0.5])
w2 = np.array([0.5,0.5])
b1=b2=0.1

y1 = np.dot(x,w1) + b1
y2 = np.dot(x,w2) + b2

print("Neuron 1 output:",y1)
print("Neuron 2 output:",y2)
```
